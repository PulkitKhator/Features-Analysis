```python
#Write the function for Multi-country ordering that takes country name as input and gives a List as output
def MultiCountryOrdering(country):
  X = df_MongoDB[(df_MongoDB['headers.country'] == country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED'))]['n'].sum(axis = 0, skipna = True)
  Y = df_MongoDB[(df_MongoDB['headers.country'] == country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED')) & (df_MongoDB['App'] == country)]['n'].sum(axis = 0, skipna = True)
  Z = X - Y
  W = (foo(Z,X))*100
  P = df_MongoDB[(df_MongoDB['headers.country'] == country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED'))]['Amount_dollar'].sum(axis = 0, skipna = True)
  Q = df_MongoDB[(df_MongoDB['headers.country'] ==country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED')) & (df_MongoDB['App'] == country)]['Amount_dollar'].sum(axis = 0, skipna = True)
  R = P - Q
  S = (foo(R,P))*100
  X2 = df_MongoDB[(df_MongoDB['App'] == country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED'))]['n'].sum(axis = 0, skipna = True)
  Y2 = df_MongoDB[(df_MongoDB['App'] ==country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED')) & (df_MongoDB['headers.country'] == country)]['n'].sum(axis = 0, skipna = True)
  Z2 = X2 - Y2
  W2 = (foo(Z2,X2))*100
  P2 = df_MongoDB[(df_MongoDB['App'] == country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED'))]['Amount_dollar'].sum(axis = 0, skipna = True)
  Q2 = df_MongoDB[(df_MongoDB['headers.country'] == country) & ((df_MongoDB['status'] == 'DELIVERED') | (df_MongoDB['status'] == 'READY') | (df_MongoDB['status'] == 'BEING_PREPARED')) & (df_MongoDB['App'] == country)]['Amount_dollar'].sum(axis = 0, skipna = True)
  R2 = P2 - Q2
  S2 = (foo(R2,P2))*100
  count = [X, Y, Z, W, P, Q, R, S, X2, Y2, Z2, W2, P2, Q2, R2, S2]
  return count
```

```python
#Write the function for OTP via email that takes bundle id(id) as input and gives a List as output
def OTP(id):
  A = len(df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'PhoneNumberVerification') & (df_OTP['EventAction'] == 'ClickOnSubmit') & (df_OTP['EventLabel'] == 'RequestOTP')])
  B = len(df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'PhoneNumberVerification') & (df_OTP['EventAction'] == 'ClickOnVerify') & (df_OTP['EventLabel'] == 'OTPVerification') & (df_OTP['OTPVerification'] == 'Success')])
  C = len(df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'PhoneNumberVerification') & (df_OTP['EventAction'] == 'ClickOnVerify') & (df_OTP['EventLabel'] == 'OTPVerification') & (df_OTP['OTPVerification'] == 'Failure')])
  D = A - B
  E = len(df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'RequestVerificationViaEmail') & (df_OTP['EventAction'] == 'ClickOnGetOTPOnEmail') & (df_OTP['EventLabel'] == 'RequestVerificationViaEmail')])
  F = len(df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'EmailIDVerification') & (df_OTP['EventAction'] == 'ClickOnVerify') & (df_OTP['EventLabel'] == 'OTPVerification')])
  G = df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'EmailIDVerification')]['Purchase'].sum(axis = 0, skipna = True)
  H = df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'EmailIDVerification')]['Amount_dollar'].sum(axis = 0, skipna = True)
  I = df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'EmailIDVerification')]['UserPurchase'].sum(axis = 0, skipna = True)
  J = df_OTP[(df_OTP['id'] == id) & (df_OTP['EventCategory'] == 'EmailIDVerification')]['Amount_dollar2'].sum(axis = 0, skipna = True)
  count = [A, B, C, D, E, F, G, H, I, J]
  return count
```
